```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Generating features
X = np.linspace(0, 10, 100)

# Adding noise noise
noise = np.random.randn(100) * 2

# Generating target values
y = 3 * X + 4 + noise
```

```python
plt.scatter(X, y, color='steelblue', alpha=0.7)
plt.title("Generated Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
def predict(X,w,b):
    return X * w + b
```

```python
def compute_loss(y_pred,y):
    return np.sum((y_pred - y)**2)
```

```python
def gradient_descent(X,y,n_iters = 1000, lr = 0.01):
    n = X.shape[0]
    w = 0
    b = 0
    loss_log = []
    print("Initializing training process")
    for i in range(n_iters):
        y_pred = predict(X,w,b)
        loss = compute_loss(y_pred,y)
        loss_log.append(loss)
        if i == 0 or (i+1) % 100 == 0:
            print(f"i = {i}\tw = {w:.6f}\tb = {b:.6f}\tloss = {loss:.6f}") 

        dw = (1 / n) * np.dot(X.T, (y_pred - y))
        db = (1 / n) * np.sum(y_pred - y)

        w -= lr * dw
        b -= lr * db
    return w,b,loss_log
```

```python
w_final, b_final, loss_log = gradient_descent(X,y)
print(f"\nfinal w = {w_final} final b = {b_final}")
```

```python
plt.plot(np.linspace(0,1000,len(loss_log)),loss_log)
plt.title("Loss over iteration")
plt.xlabel("iteration")
plt.ylabel("loss")
plt.show()
```

```python
plt.scatter(X, y, color='steelblue', alpha=0.7)
plt.plot(np.linspace(1,10,100),predict(np.linspace(1,10,100),w_final,b_final),color="red")
plt.title("Generated Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
